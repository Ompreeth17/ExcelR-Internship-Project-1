# Univariate Analysis



*   Histograms:




```python
plt.hist(data['exhaust_vacuum'], bins=20)
plt.title('Distribution of Exhaust Vacuum')
plt.xlabel('Exhaust Vacuum')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.hist(data['amb_pressure'], bins=20)
plt.title('Distribution of Ambient Pressure')
plt.xlabel('Ambient Pressure')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(data['energy_production'], bins=20)
plt.title('Distribution of Energy Production')
plt.xlabel('Energy Production')
plt.ylabel('Frequency')
plt.show()
```



*   Box plots:




```python
sns.boxplot(y=data['exhaust_vacuum'])
plt.title('Box Plot of Exhaust Vacuum')
plt.show()
```

```python
sns.boxplot(y=data['amb_pressure'])
plt.title('Box Plot of Ambient Pressure')
plt.show()
```

```python
sns.boxplot(y=data['energy_production'])
plt.title('Box Plot of Energy Production')
plt.show()
```

# Bivariate analysis

```python
plt.scatter(data['exhaust_vacuum'], data['energy_production'])
plt.title('Exhaust Vacuum vs. Energy Production')
plt.xlabel('Exhaust Vacuum')
plt.ylabel('Energy Production')
plt.show()
```

```python
plt.scatter(data['amb_pressure'], data['energy_production'])
plt.title('Ambient Pressure vs. Energy Production')
plt.xlabel('Ambient Pressure')
plt.ylabel('Energy Production')
plt.show()
```

# **Correlation matrix and heatmap:**

```python
correlation_matrix = data[['exhaust_vacuum', 'amb_pressure', 'energy_production']].corr()


sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
